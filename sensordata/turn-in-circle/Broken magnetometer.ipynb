```python
import pandas as pd 
import matplotlib.pyplot as plt

def plotVector3d(filename):
    df = pd.read_csv(
        filename,
        usecols=["value_x", "value_y", "value_z"],
        delim_whitespace=True,
        skiprows=1)
    df.plot.line()
    plt.show()

plotVector3d("magnetometer_2019_06_23_20_12_05.txt")
plotVector3d("accelerometer_2019_06_23_20_12_05.txt")
plotVector3d("gyroscope_2019_06_23_20_12_05.txt")
```
